```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd, numpy as np
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period = "max")
sp500
```

```python
sp500.index
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500.plot.line(y = "Close", use_index = True, grid = True)
```

### Will the stock go up or down

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
from sklearn.metrics import precision_score
from sklearn import metrics
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators = 100,
    min_samples_split = 100,
    random_state = 1
)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])
```

```python
# Precision = true which is predicted as true
preds = model.predict(test[predictors])
preds = pd.Series(preds, index = test.index)

print("Precision Score =", precision_score(test["Target"], preds))
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
combined.plot()
```

# Bulding a Back testing system

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined


# predictors = column names. predictors = ["Close", "Volume", "Open", "High", "Low"]
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    # 250 trading day in year
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
predictions["Target"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
metrics.confusion_matrix(
    y_true = predictions["Target"], 
    y_pred = predictions["Predictions"]
)
```

```python
horizons = [2, 5, 60, 250, 1000]   # last 2 days, last 5 days, etc..
new_predictors = []

for horizon in horizons:
    rolling_avg = sp500["Close"].rolling(horizon).mean()

    ratio_column = f"Close_Ration_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_avg

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500["Target"].shift(1).rolling(horizon).sum()

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model = RandomForestClassifier(
    n_estimators = 200,
    min_samples_split = 50,
    random_state = 1
)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]    # probabily the stock go up
    # setting 60% threshold
    preds[preds >= 0.6] = 1
    preds[preds < 0.6] = 0
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined

# predictors = column names. predictors = ["Close", "Volume", "Open", "High", "Low"]
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
metrics.confusion_matrix(
    y_true = predictions["Target"], 
    y_pred = predictions["Predictions"]
)
```
