```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period = "max")
sp500
```

```python
sp500.index
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500.plot.line(y = "Close", use_index = True, grid = True)
```

### Will the stock go up or down

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
from sklearn.metrics import precision_score
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators = 100,
    min_samples_split = 100,
    random_state = 1
)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]

model.fit(train[predictors], train["Target"])
```

```python
# Precision = true which is predicted as true
preds = model.predict(test[predictors])
preds = pd.Series(preds, index = test.index)

print("Precision Score =", precision_score(test["Target"], preds))
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
combined.plot()
```

# Bulding a Back testing system

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```
