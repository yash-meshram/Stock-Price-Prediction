```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
from datetime import datetime

# Sample data with dates
dates = [datetime(2023, 1, 1), datetime(2023, 1, 2), datetime(2023, 1, 3), datetime(2023, 1, 4), datetime(2023, 1, 5)]
x = date2num(dates)
y = np.array([2, 3, 4, 3.5, 5])

# Perform linear regression
coefficients = np.polyfit(x, y, 1)
line = np.polyval(coefficients, x)

# Plot the scatter plot
plt.scatter(dates, y, label='Data Points')

# Plot the linear line
plt.plot(dates, line, color='red', label=f'Linear Fit: {coefficients[0]:.2f}x + {coefficients[1]:.2f}')

# Beautify the plot
plt.xlabel('Date')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Linear Fit')
plt.legend()
plt.show()
```
