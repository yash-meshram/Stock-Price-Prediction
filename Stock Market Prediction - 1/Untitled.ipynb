```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap

# Sample DataFrame
data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04']),
    'Values': [5, -3, 8, -2]
}

df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

# Separate positive and negative values
positive_values = df[df['Values'] >= 0]
negative_values = df[df['Values'] < 0]

# Convert 'Date' values to numerical format
date_numeric = range(len(df))
date_numeric_positive = range(len(positive_values))

# Plotting the graph with a single continuous line
segments = [((x1, y1), (x2, y2)) for x1, y1, x2, y2 in zip(date_numeric_positive, positive_values['Values'], date_numeric, negative_values['Values'])]
colors = ['green' if value >= 0 else 'red' for value in df['Values']]

# Create a LineCollection
lc = LineCollection(segments, colors=colors, linewidth=2)

# Plot the LineCollection
fig, ax = plt.subplots()
ax.add_collection(lc)

# Beautify the plot
ax.autoscale()
ax.margins(0.1)
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Line Graph with Colors Based on Positive/Negative Values')
plt.show()
```
